```r
df<-data.frame(
student_id = c(101,102,103,104,105,106,107,108),
math = c(75,88,92,67,81,95,78,84),
programming = c(80,90,89,70,75,98,82,88)
)
df
```

```r
#Calculate the average marks for each subject.
math_avg<-mean(math)
math_avg
prog_avg<-mean(programming)
prog_avg
```

```r
# Find students scoring above the class average in Programming.
df[df$programming>prog_avg,]
```

```r
# Calculate the total marks of each student.
df$total<-math+programming
df
```

```r
df$grade<-ifelse(df$total>=180,'A',ifelse((df$total>=160),"B",'C'))
df
```
