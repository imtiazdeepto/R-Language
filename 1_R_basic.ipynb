```r
#create database
df<-data.frame(
    name=c('A','B','C','D'),
    Id=c(1,2,3,4),
    course=c("Math","English","bangla","Math"),
    marks=c(20,30,50,50)

)
df
```

```r
pie(df$marks,labels=df$course,main = "marks by course")
```

```r
df[5,]<-list("E",5,"English",80)
df[6,]<-list("F",6,"English",90)
df
```

```r
df$Dept<-c("SWE","SWE","CSE","CSE","CIS","CIS")
```

```r
df
```

```r
#Dept wise mark avg
aggregate(marks~Dept,data=df,mean)
```

```r
df[which.max(df$marks),5]
```

```r
df[which.max(df$marks),'Dept']
```

```r
# Create a bar plot for each student's marks
barplot(df$marks,
        names.arg=df$name,
        col = "skyblue",
        main = "Marks for Each Student",
        xlab = "Student Name",
        ylab = "Marks",
        ylim = c(0, 100)) # Sets the Y-axis limit from 0 to 100
```

```r
# Histogram of marks
hist(df$marks,
     col = "lightgreen",
     main = "Distribution of Marks",
     xlab = "Marks Range",
     ylab = "Number of Students") # adjusts the number of bars
```

```r
boxplot(df$marks,
        col = "tomato",
        main = "Overall Marks Boxplot",
        ylab = "Marks")
```

```r
# Standard index plot of marks
plot(df$Id,df$marks,
     type = "b",          # "b" means BOTH lines and points
     pch = 19,            # Solid circle symbol for points
     col = "blue",
     main = "Student Marks Trend",
     xlab = "Student Index (1 to 6)",
     ylab = "Marks")          # Turn off default X axis to customize it below

```
